##KNN 알고리즘
k-NN: k-최근접 이웃k-Nearest Neighbor의 약자 다수결 방식으로 데이터의 레이블을 할당하는 분류 방식.

```python
#matplot 라이브러리 이용한 데이터 시각화
#• 닥트훈트: 빨간색 동그라미, 진돗개 파란색 세모 표시
#• 가로축=길이, 세로축=높이

import matplotlib.pyplot as plt

# 닥스훈트의 길이와 높이 데이터
dach_length = [55, 57, 64, 63, 58, 49, 54, 61]
dach_height = [30, 31, 36, 30, 33, 25, 37, 34]
# 진돗개의 길이와 높이 데이터
jin_length = [56, 47, 56, 46, 49, 53, 52, 48]
jin_height = [52, 52, 50, 53, 50, 53, 49, 54]

plt.scatter(dach_length, dach_height, c ='r', label='Dachshund')
plt.scatter(jin_length, jin_height, c='b', marker='^', label='jindo dog')

# 새로운 데이터를 추가해 확인하기 길이 59, 높이 35

newdata_length = [59]
newdata_height = [35]

plt.scatter(newdata_length, newdata_height, s=100, marker ='p', 
            c='g', label='newdata')


plt.xlabel('Length')
plt.ylabel('Height')
plt.title('Dog size')
plt.legend(loc='upper right')
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

d_data = np.column_stack((dach_length, dach_height)) #닥스훈트의 특성값 배열화
d_label = np.zeros(len(d_data)) # 닥스훈트는 0으로 레이블링(정답 달기)

j_data = np.column_stack((jin_length, jin_height)) # 진돗개의 특성값 배열화
j_label = np.ones(len(j_data)) # 진돗개는 1로 레이블링
# ----여기까지가 데이터 정리 단계----

newdata = [[59, 35]]

dogs = np.concatenate((d_data, j_data)) # 데이터 연결 시키기 d_data 다음 j_data가 나옴
labels = np.concatenate((d_label, j_label)) # 정답도 똑같이 연결

dog_classes = {0:'닥스훈트', 1:'진돗개'} 
k = 3 # k를 3으로 하는 kNN 분류기
knn = KNeighborsClassifier(n_neighbors = k) #KNN 알고리즘 사용

#데이터와 라벨을 모델에 입력, 훈련 시키기
knn.fit(dogs, labels)
y_pred = knn.predict(newdata) # 판정 결과를 넣는다 (1개)
print('데이터', newdata, ', 판정 결과:', dog_classes[y_pred[0]])
```
